```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Dados das vendas anuais da Red Bull
dados = {
    'Ano': list(range(2011, 2025)),
    'Latas Vendidas (bilhões)': [
        4.6, 5.2, 5.3, 5.6, 5.9, 6.1, 6.3, 6.8,
        7.5, 7.9, 9.8, 10.4, 12.0, 12.7
    ],
    'Receita (bilhões de euros)': [
        4.25, 4.93, 5.3, 4.93, 6.1, 6.3, 6.6, 6.9,
        6.1, 6.3, 7.1, 7.4, 10.55, 11.23
    ]
}

# Criar DataFrame
df = pd.DataFrame(dados)


# Configurar o estilo do gráfico
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(14, 7))

# Definir a largura das barras
largura = 0.4
x = np.arange(len(df['Ano']))

# Criar as barras
ax.bar(x - largura/2, df['Latas Vendidas (bilhões)'], width=largura, label='Latas Vendidas', color='skyblue')
ax.bar(x + largura/2, df['Receita (bilhões de euros)'], width=largura, label='Receita', color='salmon')

# Adicionar títulos e rótulos
ax.set_title('Vendas Anuais da Red Bull (2011–2024)', fontsize=16)
ax.set_xlabel('Ano', fontsize=12)
ax.set_ylabel('Bilhões', fontsize=12)
ax.set_xticks(x)
ax.set_xticklabels(df['Ano'], rotation=45)
ax.legend()
plt.tight_layout()

# Exibir o gráfico
plt.show()
```

```python
#Este tipo de análise é útil quando você deseja entender a distribuição percentual de uma variável categórica em relação a outra. Por exemplo:

#Analisar como as vendas variaram ao longo dos anos para diferentes quantidades de produtos vendidos.

#Entender a distribuição de categorias em diferentes períodos ou grupos
pd.crosstab(df['Latas Vendidas (bilhões)'], df['Ano'], normalize='index') * 100
```

```python
df["lucro"]=df["Receita (bilhões de euros)"]*df["Latas Vendidas (bilhões)"]
df.head(100)
```

```python

# Supondo que você já tenha um DataFrame 'df' com as colunas:
# 'Ano', 'Latas Vendidas (bilhões)', 'Receita (bilhões de euros)', 'lucro'

# 1. Agrupar os dados por 'Ano', somando os valores das colunas selecionadas
colunas = ["Latas Vendidas (bilhões)", "Receita (bilhões de euros)", "lucro"]
df_geral = df.groupby("Ano")[colunas].sum().round(2)

# 2. Calcular a margem de lucro para cada ano
#    Margem de lucro (%) = (lucro / receita) * 100
df_geral['margem_lucro (%)'] = (
    df_geral['lucro'] / df_geral['Receita (bilhões de euros)'] * 100
).round(2)

# 3. Exibir o DataFrame resultante
print(df_geral)

```

```python
import matplotlib.pyplot as plt

# Gráfico de barras da margem de lucro por ano
df_geral['margem_lucro (%)'].plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.title('Margem de Lucro (%) por Ano')
plt.xlabel('Ano')
plt.ylabel('Margem de Lucro (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
